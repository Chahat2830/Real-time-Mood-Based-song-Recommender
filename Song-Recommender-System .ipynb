```python
features = [
    'acousticness','danceability','energy','instrumentalness',
    'liveness','loudness','speechiness','tempo','valence'
]
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[features])
```

```python
def recommend_songs(song_name, df, scaled_features, top_n=6):
    if song_name not in df['track_name'].values:
        return []

    # Find index
    index = df[df['track_name'] == song_name].index[0]

    # Compute similarity only for this song (1 x N)
    sim_scores = cosine_similarity(
        scaled_features[index].reshape(1, -1),
        scaled_features
    )[0]

    # Get top matches
    top_indices = sim_scores.argsort()[::-1][1:top_n+1]

    # Return rows
    return df.iloc[top_indices]
```

```python
# 3) Mood mapping -> filter function
# You can adjust thresholds as you prefer
def filter_by_mood(df_subset, mood):
    if mood is None:
        return df_subset
    mood = mood.lower()
    if mood == "happy":
        return df_subset[(df_subset["valence"] > 0.6) & (df_subset["danceability"] > 0.4)]
    elif mood == "sad":
        return df_subset[df_subset["valence"] < 0.4]
    elif mood == "energetic":
        return df_subset[(df_subset["energy"] > 0.65) | (df_subset["tempo"] > 120)]
    elif mood == "calm":
        return df_subset[(df_subset["energy"] < 0.45) & (df_subset["acousticness"] > 0.4)]
    else:
        return df_subset
```

```python
# 4) Hybrid function: content-based + mood filter
def hybrid_recommend(song_name, df, scaled_features, mood=None, top_n=6):
    base_recs = recommend_songs(song_name, df, scaled_features, top_n=50)  # get larger pool
    if base_recs.empty:
        return base_recs
    mood_filtered = filter_by_mood(base_recs, mood)
    # If mood filtering reduces too much, fallback to base_recs
    if len(mood_filtered) < top_n:
        # fill remaining from base_recs
        remaining = base_recs[~base_recs.index.isin(mood_filtered.index)].head(top_n - len(mood_filtered))
        result = pd.concat([mood_filtered, remaining]).head(top_n)
    else:
        result = mood_filtered.head(top_n)
    return result.reset_index(drop=True)
```

```python
# 5) Quick tests
song_example = df['track_name'].iloc[10]   # change to a real name if you want
print("Selected:", song_example)

print("\n-- Content-based recommendations --")
print(recommend_songs(song_example, df, scaled_features, top_n=6)[["track_name","artist_name","valence","energy"]])

print("\n-- Hybrid (mood=Happy) recommendations --")
print(hybrid_recommend(song_example, df, scaled_features, mood="Happy", top_n=6)[["track_name","artist_name","valence","energy"]])
```
